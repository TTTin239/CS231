#Xác định và hiện thị các điểm interest/key points của các cặp ảnh

```python
def Harris(path_filename1, path_filename2):
  img1 = cv.imread(path_filename1)
  operatedImage1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
  operatedImage1 = np.float32(operatedImage1)
  dest1 = cv.cornerHarris(operatedImage1, 2, 5, 0.07)
  dest1 = cv.dilate(dest1, None)
  img1[dest1 > 0.01 * dest1.max()] = [0, 0, 255]

  img2 = cv.imread(path_filename2)
  operatedImage2 = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)
  operatedImage2 = np.float32(operatedImage2)
  dest2 = cv.cornerHarris(operatedImage2, 2, 5, 0.07)
  dest2 = cv.dilate(dest2, None)
  img2[dest2 > 0.01 * dest2.max()] = [0, 0, 255]
  return img1, img2
```

```python
path_cow1 = os.path.join(dir_images, 'cow1.jpg')
path_cow2 = os.path.join(dir_images, 'cow2.jpg')

path_match1 = os.path.join(dir_images, 'match1.jpg')
path_match2 = os.path.join(dir_images, 'match2.jpg')

path_graf_img1 = os.path.join(dir_images, 'graf_img1.jpg')
path_graf_img5 = os.path.join(dir_images, 'graf_img5.jpg')
```

##cow1.jpg và cow2.jpg

```python
img1, img2 = Harris(path_cow1, path_cow2)
cv2_imshow(img1)
cv2_imshow(img2)
```

##match1.jpg và match2.jpg

```python
img3, img4 = Harris(path_match1, path_match2)
cv2_imshow(img3)
cv2_imshow(img4)
```

##graf_img1.jpg và graf_img5.jpg

```python
img5, img6 = Harris(path_graf_img1, path_graf_img5)
cv2_imshow(img5)
cv2_imshow(img6)
```
